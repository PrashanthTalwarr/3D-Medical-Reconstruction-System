```python
# ============================================================================
# CELL 9: Test Inference on New Sample
# ============================================================================

print("\n" + "="*70)
print("üß™ TESTING INFERENCE ON NEW SAMPLE")
print("="*70 + "\n")

import torch
from local.model import UNet3D
from local.diffusion import DDPM3D
from local.dataset import SyntheticMedical3D
import matplotlib.pyplot as plt

device = torch.device('cuda')

# Load trained model
print("üîÑ Loading trained model...")
model = UNet3D(base_channels=32).to(device)
model.load_state_dict(torch.load('checkpoints/final_model.pt'))
model.eval()
print("‚úÖ Model loaded!\n")

# Initialize diffusion
diffusion = DDPM3D(timesteps=200, device=device)

# Create new test sample
print("üîÑ Creating test sample...")
test_dataset = SyntheticMedical3D(num_samples=1, volume_size=32, augment=False)
test_volume = test_dataset[0].unsqueeze(0).to(device)

# Add noise
t = torch.full((1,), diffusion.timesteps // 2, device=device).long()
noisy_volume, _ = diffusion.q_sample(test_volume, t)

# Reconstruct
print("üîÑ Running inference (this may take a moment)...\n")
with torch.no_grad():
    reconstructed = diffusion.sample(model, test_volume.shape, device)

reconstructed = torch.clamp(reconstructed, 0, 1)

# Calculate metrics
from local.utils import evaluate_reconstruction
metrics = evaluate_reconstruction(reconstructed, test_volume)

print("Inference Results:")
print("-" * 40)
for metric, value in metrics.items():
    print(f"  {metric:12s}: {value:.4f}")
print("-" * 40 + "\n")

# Visualize
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

volumes = [test_volume, noisy_volume, reconstructed]
titles = ['Original', 'Noisy Input', 'Reconstructed']

for ax, vol, title in zip(axes, volumes, titles):
    vol_np = vol.cpu().numpy()[0, 0]
    mid = vol_np.shape[0] // 2
    ax.imshow(vol_np[mid, :, :], cmap='gray', vmin=0, vmax=1)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.savefig('results/inference_test.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n‚úÖ Inference complete!")
print("‚úÖ Saved: results/inference_test.png")
```
